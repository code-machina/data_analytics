## pandas 를 이용한 표준시간대 세어보기

````python
from pandas import DataFrame, Series
import pandas as pd
````

`DataFrame().info()` 메서드는 칼럼의 이름 목록 그리고 각 칼럼당 데이터 개수 마지막으로 데이터 타입을 지정한다.
- Q. 그렇다면, 칼럼의 데이터 타입은 어떻게 변환할까?

````python
frame = DataFrame(records)
frame.info()
````

DataFrame 에서 `[]` 연산자를 통해 특정 칼럼의 값을 아래와 같이 가져오는 경우, Series 객체이다.

````python
frame['tz'][:10]
````

Series 객체는 value_counts 메서드를 지원하며 해당 메서드는 칼럼 내 데이터의 카운트를 측정하여 알려준다.
그러나, 공백을 포함하므로 없는 값에 대해서는 미리 fillna 와 불리언 배열 색인(Boolean array indexing)을 통해 대체하는 처리 작업을 해준다.

````python
tz_counts = frame['tz'].value_counts()
tz_counts[:10]
````

````python
clean_tz = frame['tz'].fillna('Missing') # 없는 값을 대체
clean_tz[clean_tz == ''] = 'Unknown' # 비어 있는 값은 불리언 배열 색인(Boolean array indexing)을 통해 처리한다.
# 다시 카운트를 뽑아본다.
tz_counts = clean_tz.value_counts()
tz_counts[:10]
````

처리 작업을 통해 비어있는 값을 처리할 수 있었다. Missing 과 Unknown 이 각각 521, 120 개 있음을 알 수 있다.

## 브라우저 정보 분석하기

tz 값을 통해 접속 위치를 유추할 수 있으며 브라우저 정보를 결합하여 그래프로 나타낼 수 있다.

````python
results = Series([x.split()[0].strip() for x in frame.a.dropna()])
results[:5]
````

````python
results.value_counts()[:8]
````

책에서는 배열의 맨앞을 잘라 내었지만 아래의 리스트를 보면 Firefox, Safari 등의 키워드가 보인다.
중간에 Chrome 키워드가 있는 것도 보이며 두가지 키워드가 결합되어 나타나는 경우가 있다.  

아래의 URL 에서 브라우저에 대한 정확한 정보를 구할 수 있다.
https://developers.whatismybrowser.com/useragents/explore/software_name/comodo-dragon/ 

````python
frame.a.dropna().value_counts()[:10]
# Opera 키워드 찾아보기
# Dataframe 에서 Brwoser 칼럼을 뽑아낸다. 
browsers = frame[frame.a.notnull()].a
# 불리언 배열 색인을 통해 Opera 키워드를 포함하는 리스트를 추출한다.
opera_browsers = browsers[browsers.str.contains('Opera')]
# 오페라 브라우저의 리스트를 추출한다.
opera_browsers.value_counts()
# frame.a.dropna()[frame.a.notnull().str.contains('Opera')][:5]
````

위의 데이터에서 `Opera/9.80 ()` 의 괄호 안의 문자열이 운영체제 정보인 것으로 보인다. 정규표현식을 통해 괄호의 내용만 가져올 수 있을까?  
스택 오버플로우에 관련 내용을 찾을 수 있었다.  
https://stackoverflow.com/questions/18219030/how-to-extract-the-content-between-two-brackets-by-using-grep
아래는 더욱 정확한 사용 방법을 알 수 있다. (extract 메서드 사용)
https://stackoverflow.com/questions/16842001/copy-text-between-parentheses-in-pandas-dataframe-column-into-another-column
```text
Still using grep and regex

grep -oP '\(\K[^\)]+' file
\K means that use look around regex advanced feature. More precisely, it's a positive look-behind assertion, you can do it like this too :

grep -oP '(?<=\()[^\)]+' file
if you lack the -P option, you can do this with perl :

perl -lne '/\(\K[^\)]+/ and print $&' file
Another simpler approach using awk

awk -F'[()]' '{print $2}' file
```
  
- 아래의 예시는 괄호(parenthesis) 내의 내용을 제거하는 것이다.
  
```python
energy['Country'] = energy['Country'].str.replace(r"\s+\(.*\)","")
```
  
- 아래의 예시가 우리가 원하는 내용이다.
  
```python
s.str.extract('.*\((.*)\).*')
```

몇 번의 시도 끝에 Opera 브라우저 내에서 운영체제 정보를 구체적으로 추출하는데 성공하였다.

````python
opera_browsers.str.extract('.*\((.*)\).*', expand=False)
# 그러나 browser language 정보가 붙어있으므로 이를 조금 수정해 본다.
opera_browsers.str.extract('\((.*)\)', expand=False).str.split(';', expand=True)[0].value_counts()
````

## 특정 필드들로 데이터 그룹화 하기

위에서 추출하는 방법을 확인하였으며 이를 바탕으로 그룹화를 하기 전에 새로운 칼럼을 생성해본다. 새로운 칼럼은 브라우저 정보만을 추출한 정보이다.

````python
frame["os"] = frame.a.str.extract('\((.*?)\)', expand=False).str.split(';', expand=True)[0]
frame["os"]
````

````python
top_5_os = frame.os.value_counts()[:5]
````

````python
list(top_5_os)
# frame.os.isin(list(top_5_os)).value_counts()
````

````python
by_tz_os = frame.groupby(['tz', 'os'])
````

````python
by_tz_os['os']
````

````python
agg_counts = by_tz_os.size().unstack().fillna(0)
agg_counts.head(10)
````

````python
agg_counts[:10]
````

````python
indexer = agg_counts.sum(1).argsort()
indexer[:10]
````

````python
count_subset = agg_counts.take(indexer)[-10:]
count_subset
````

````python
count_subset.plot(kind='barh', stacked=True)
````
